### Cleaning Data

```python
#Mengisi data tipe numerik
main_df.interpolate(method='linear', limit_direction='both', inplace=True)

#Mengisi data tipe kategorikal
main_df['wd'] = main_df['wd'].fillna(method='ffill')
```

```python
main_df.isna().sum()
```

```python
#Menangani kondisi oulier
numerical_cols = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

for col in numerical_cols:
  Q1 = (main_df[col]).quantile(0.25)
  Q3 = (main_df[col]).quantile(0.75)
  IQR = Q3 - Q1

  maximum = Q3 + (1.5*IQR)
  minimum = Q1 - (1.5*IQR)

  kondisi_lower_than = main_df[col] < minimum
  kondisi_more_than = main_df[col] > maximum

  main_df.loc[kondisi_more_than, col] = maximum
  main_df.loc[kondisi_lower_than, col] = minimum
```

```python
for col in numerical_cols:
  plt.figure(figsize=(10, 5))
  sns.boxplot(x=main_df[col])
  plt.title(f'Boxplot {col}')
  plt.ylabel(col)
  plt.show()
```

**Insight:**
- Pada tahap ini dilakukan proses penanganan missing data dengan menerapkan teknik interpolasi dengan metode both untuk data numerik sedangkan untuk data kategorikal dilakukan dengan cara mengisi dengan metode forward fill (ffill) yaitu mengisi data berdasarkan data sebelumnya hal ini dilakukan karena data kategorikal yang memiliki nilai kosong (wd) merupakan data arah mata angin sehingga diasumsikan arah mata angin tidak banyak berubah dalam waktu singkat.
- Sedangkan untuk proses penanganan data outlier, karena data outlier begitu banyak dilakukan proses imputation dengan cara mengganti nilai pada data outlier menjadi nilai batas. Hal ini dilakukan agar tidak terlalu banyak data yang dibuang yang dapat merusak struktur data.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Stasiun manakah yang memiliki catatan rata rata kualitas udara terburuk berdasarkan rata-rata kadar PM2.5 dan PM10?

```python
df_meanstation = main_df.groupby('station')[['PM2.5','PM10']].mean()
```

```python
plt.figure(figsize=(10, 5))
df_meanstation_sorted = df_meanstation['PM2.5'].sort_values(ascending=True)
ax = df_meanstation_sorted.plot(kind='barh', color='salmon')

for i, bar in enumerate(ax.patches):
    value = df_meanstation_sorted.iloc[i]

    if i == len(ax.patches) - 1:
        bar.set_color('red')

    ax.text(value + 0.5, bar.get_y() + bar.get_height()/2,
            f'{value:.2f}',
            va='center', ha='left', fontsize=10)

plt.xlabel('Rata-rata kadar PM2.5')
plt.ylabel('Stasiun')
plt.title('Rata-rata kadar PM2.5 per Stasiun')
plt.xlim(0, df_meanstation_sorted.max() + 10)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df_meanstation_sorted = df_meanstation['PM10'].sort_values(ascending=True)
ax = df_meanstation_sorted.plot(kind='barh', color='salmon')

for i, bar in enumerate(ax.patches):
    value = df_meanstation_sorted.iloc[i]

    if i == len(ax.patches) - 1:
        bar.set_color('red')

    ax.text(value + 0.5, bar.get_y() + bar.get_height()/2,
            f'{value:.2f}',
            va='center', ha='left', fontsize=10)

plt.xlabel('Rata-rata kadar PM10')
plt.ylabel('Stasiun')
plt.title('Rata-rata kadar PM10 per Stasiun')
plt.xlim(0, df_meanstation_sorted.max() + 10)
plt.tight_layout()
plt.show()
```

**Insight**:
- Berdasarkan visualisasi di atas, kualitas udara terburuk berdasarkan kadar PM2.5 adalah pada stasiun Dongsi (81.30), lalu untuk kualitas udara terbaik adalah pada stasiun Dingling (64.47).
- Sementara itu, kualitas udara terburuk berdasarkan kadar PM10 adalah pada stasiun Gucheng (115.30), lalu untuk kualitas udara terbaik adalah pada stasiun Dingling (82.37).
- Stasiun Dingling merupakan wilayah dengan kualitas udara terbaik secara keseluruhan karena memiliki kadar PM2.5 dan PM10 terendah.
